```python
new_rows_dict = {
    'TIME': [2000,2000,2000],
    'GEO': ['USA', 'Canada', 'Uganda'],
    'Value': [5, 4.18, 5.15]
}

new_rows_df = pd.DataFrame(new_rows_dict)

edu_new_rows = edu.append(new_rows_df, ignore_index = True, sort=False)

final_row_dict = {
    'TIME' : [2021],
    'GEO' : ['Mars'],
    'Value' : [1000]
}

final_row_df = pd.DataFrame(final_row_dict)

edu_final_row = edu_new_rows.append(final_row_df, ignore_index = True, sort = False)
edu_final_row
```
