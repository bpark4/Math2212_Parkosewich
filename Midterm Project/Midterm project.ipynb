```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
cen = pd.read_csv("census.csv")
cen
```

```python
cen.shape
```

```python
orig_states = cen[['state_fips_code']]
states = orig_states.drop_duplicates()

state_names = []
state_tfi = []

for i in range(0,47):
    cen2 = cen[cen['state_fips_code'] == states['state_fips_code'][states.index[i]]][['total_family_income']]
    mean = cen2['total_family_income'].mean()
    state_names.append(states['state_fips_code'][states.index[i]])
    state_tfi.append(mean)
    

plt.figure(figsize=(15,5))
plt.bar(state_names,state_tfi)
plt.suptitle("Mean Total Family Income by State")
```
