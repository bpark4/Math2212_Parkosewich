```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# This code gets the data from the census.csv file and puts it into a dataframe

```python
cen = pd.read_csv("census.csv")
cen
```

## Shows the size of the data set, in (rows, columns)

```python
cen.shape
```

# This plot zooms in on the states which have a mean family income over 70000 dollars

# Allows for easy and more precise comparison of states within this range

```python
# get state names
orig_states = cen[['state_fips_code']]
states = orig_states.drop_duplicates()

# state name and totalfamilyincome lists
state_names = []
state_names2 = []
state_names3 = []
state_names4 = []
state_tfi = []
state_tfi2 = []
state_tfi3 = []
state_tfi4 = []

# gets mean tfi of each state, seperates them by amount to show state names
for i in range(0,47):
    cen2 = cen[cen['state_fips_code'] == states['state_fips_code'][states.index[i]]][['total_family_income']]
    mean = cen2['total_family_income'].mean()
    if (mean > 70000):
        state_names.append(states['state_fips_code'][states.index[i]])
        state_tfi.append(mean)
    else:
        if ((mean < 70000) & (mean > 50000)):
            state_names2.append(states['state_fips_code'][states.index[i]])
            state_tfi2.append(mean)
        else:
            if ((mean < 50000) & (mean > 35000)):
                state_names3.append(states['state_fips_code'][states.index[i]])
                state_tfi3.append(mean)
            else:
                state_names4.append(states['state_fips_code'][states.index[i]])
                state_tfi4.append(mean)
    

# plot data
plt.figure(figsize=(15,5))
plt.bar(state_names,state_tfi)
plt.suptitle("Mean Total Family Income (if over $70000) by State")
```

# This plot zooms in on the states which have a mean family income between 50000 and 70000 dollars

# Allows for easy and more precise comparison of states within this range

```python

# plot data
plt.figure(figsize=(15,5))
plt.bar(state_names2,state_tfi2)
plt.suptitle("Mean Total Family Income (between $50000 and $70000) by State")
```

# This plot zooms in on the states which have a mean family income between 35000 and 50000 dollars

# Allows for easy and more precise comparison of states within this range

```python

# plot data
plt.figure(figsize=(15,5))
plt.bar(state_names3,state_tfi3)
plt.suptitle("Mean Total Family Income (between $35000 and $50000) by State")
```

# Graph that sorts mean personal income into age groups

# Shows how age effects income, relatively normally distributed

## Someone who is younger has less experience, thus gets paid less
## Someone who is much older is likely in retirement
## In the mid-range ages people are more likely to have a job that they are experienced in, and thus get paid more for it

```python
# sorts census by age
ages = cen.sort_values(by = ['age'], ascending = True, inplace = False)

# gets mean income for various age brackets
under16 = ages[ages['age'] < 16][['total_personal_income']].mean()[0]
b15and26 = ages[(ages['age'] < 26) & (ages['age'] > 15)][['total_personal_income']].mean()[0]
b25and36 = ages[(ages['age'] < 36) & (ages['age'] > 25)][['total_personal_income']].mean()[0]
b35and46 = ages[(ages['age'] < 46) & (ages['age'] > 35)][['total_personal_income']].mean()[0]
b45and56 = ages[(ages['age'] < 56) & (ages['age'] > 45)][['total_personal_income']].mean()[0]
b55and66 = ages[(ages['age'] < 66) & (ages['age'] > 55)][['total_personal_income']].mean()[0]
b65and76 = ages[(ages['age'] < 76) & (ages['age'] > 65)][['total_personal_income']].mean()[0]
b75and86 = ages[(ages['age'] < 86) & (ages['age'] > 75)][['total_personal_income']].mean()[0]
over85 = ages[ages['age'] > 85][['total_personal_income']].mean()[0]

# list of names for age brackets
cat = ["age<16","15<age<26","25<age<36","35<age<46","45<age<56","55<age<66","65<age<76","75<age<86","age>85"]

# list of mean personal income of age brackets
age_tpi = [under16,b15and26,b25and36,b35and46,b45and56,b55and66,b65and76,b75and86,over85]

# plot data
plt.figure(figsize=(15,5))
plt.bar(cat,age_tpi)
plt.suptitle("Mean personal income by age")
```
