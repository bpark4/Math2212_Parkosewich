#### ADDING ROWS TO A DATAFRAME

The idea is to make the collection of new rows into a new dataframe. Then you just "glue" that new dataframe onto the end of your original dataframe.

```python
# to create new dataframe, start with a dictionary

new_rows_dict = {
    'TIME': [2000,2000,2000],
    'GEO': ['USA', 'Canada', 'Uganda'],
    'Value': [5, 4.18, 5.15]
}

# turns dictionary into dataframe
new_rows_df = pd.DataFrame(new_rows_dict)
new_rows_df
```

```python
# Append new_rows_df to the end of edu

edu_new_rows = edu.append(new_rows_df, ignore_index = True, sort=False)

edu_new_rows
```

```python
# Worksheet four

custom_rows = {
    'TIME' : [2021],
    'GEO' : ['New Haven'],
    'Value' : [5.32]
}

custom_df = pd.DataFrame(custom_rows)

edu_new_custom = edu_new_rows.append(custom_df, ignore_index = True, sort = False)
edu_new_custom.tail()
```

## Filtering Dataframes

Boolean Filtering

```python
# return rows with Value > 4
edu[ edu['Value'] > 4 ]
```

```python
# return the rows with value between 4 and 6

edu[ (edu['Value'] > 4) & (edu['Value'] < 6)]
```

```python
# return rows for years 2002 and 2003

edu[(edu['TIME'] == 2002) | (edu['TIME'] == 2003)]
```

```python
# Filter some of the columns
edu[edu['TIME'] == 2002][['GEO', 'Value']]
```

```python
# return Value between 5 and 7 of year 2003
edu[(edu['Value'] > 5) & (edu['Value'] < 7) & (edu['TIME'] == 2003)][['GEO']]
```

```python
# return only rows with missing Value data
edu[ edu['Value'].isnull() ]

# return the rows with NO missing data
edu[ (edu['Value'].notnull()) | (edu['GEO'].notnull())]
```

# Aggregating Data

how can we sum, average, or otherwise aggregate data that shares a similar property?

*Standard aggregating functions from numpy*

count() -- this function counts how many values there are

sum() -- adds the values of interest

mean() -- finds the mean of the aggregate values

median()

min() -- finds minimum value

max() -- finds maximum value

prod() -- multiplies all values

std() -- returns the standard deviation of all the values

var() -- returns the variance of all the values

```python
# first column is what you group by, second is what to aggregate by
group_mean = edu[['GEO','Value']].groupby('GEO').mean() # finds mean value within each country
print group_mean
```

```python
group_max = edu[['TIME','Value']].groupby('TIME').min()
print group_max
```

# Pivot Tables

A pivot table is similar to these tables that we got using .groupby() except that instead of aggregating over one variable we aggregate over two variables

```python
ptable = pd.pivot_table(
    edu,
    values = 'Value',
    index = 'TIME',
    columns = 'GEO',
    aggfunc = np.sum
)
ptable
```

```python
# worksheet 6
ptable2 = pd.pivot_table(
    edu,
    values = 'Value',
    index = 'GEO',
    columns = 'TIME',
    aggfunc = np.mean
)

ptable2
```
