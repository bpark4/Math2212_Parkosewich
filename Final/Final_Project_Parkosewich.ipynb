# Imports all necessary libraries as well as the census dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

cen = pd.read_csv("census.csv")
cen
```

# Cleans the census dataframe with respect to family income
## Age does not need to be cleaned as each entry has an age associated with it

```python
cen_cleaned_tfi = cen[cen['total_family_income'].notnull()]
cen_cleaned_tfi
```

## Creates the 4 dataframes to be used in calculations

```python
cali_tex_tfi = cen_cleaned_tfi[(cen_cleaned_tfi['state_fips_code']=='California')|(cen_cleaned_tfi['state_fips_code']=='Texas')]
range_tfi = cen_cleaned_tfi
florida_ny_age_avg = cen[(cen['state_fips_code']=='Florida')|(cen['state_fips_code']=='New York')]
florida_ny_age_std = cen[(cen['state_fips_code']=='Florida')|(cen['state_fips_code']=='New York')]
```

# IQR function used in dropping outliers of the California and Texas median comparison
## Dropped values using IQR since the problem is asking for median values of datasets

```python
def IQR(x):
    return x.quantile(.75)-x.quantile(.25)
```

# Splits California and Texas to compare datasets and shows histograms
## Texas shows generally lower family incomes and less people
### Drop the outliers due to a few obvious outliers within the dataset
### The 5 highest family incomes in each group were dropped

```python
cali_tfi = cali_tex_tfi[cali_tex_tfi['state_fips_code']=='California']
tex_tfi = cali_tex_tfi[cali_tex_tfi['state_fips_code']=='Texas']

cali_tfi_sug = cali_tfi['total_family_income']
cali_tfi_median = cali_tfi['total_family_income'].median()
cali_tfi_iqr = IQR(cali_tfi['total_family_income'])
tex_tfi_sug = tex_tfi['total_family_income']
tex_tfi_median = tex_tfi['total_family_income'].median()
tex_tfi_iqr = IQR(tex_tfi['total_family_income'])

cali_tfi = cali_tfi.drop(
    cali_tfi.index[
        (cali_tfi_sug > cali_tfi_median + (1.5 * cali_tfi_iqr)) | (cali_tfi_sug < cali_tfi_median - (1.5 * cali_tfi_iqr) )
    ]
)
tex_tfi = tex_tfi.drop(
    tex_tfi.index[
        (tex_tfi_sug > tex_tfi_median + (1.5 * tex_tfi_iqr)) | (tex_tfi_sug < tex_tfi_median - (1.5 * tex_tfi_iqr) )
    ]
)

cali_tfi['total_family_income'].hist()
tex_tfi['total_family_income'].hist()
plt.show()
```

## California has more people in their dataset, potentially causing a higher probability of high income earners

```python
print "California family income mean: ", cali_tfi['total_family_income'].mean()
print "California family income standard deviation: ", cali_tfi['total_family_income'].std()
print "California family income median: ", cali_tfi['total_family_income'].median()
print "\nTexas family income mean: ", tex_tfi['total_family_income'].mean()
print "Texas family income standard deviation: ", tex_tfi['total_family_income'].std()
print "Texas family income median: ", tex_tfi['total_family_income'].median()
```

### Bootstrapping function for finding the difference in median between the two states. Gets a more accurate representation of the whole population

```python
def median_diff_bootstrap(sample_data1, sample_data2, number_of_resamples):
    x = [0]*number_of_resamples
    for i in range(number_of_resamples):
        resample1 = [
            sample_data1[j]
            for j 
            in np.random.randint(len(sample_data1), size=len(sample_data1))
        ]
        resample2 = [
            sample_data2[j]
            for j 
            in np.random.randint(len(sample_data2), size=len(sample_data2))
        ]
        x[i] = np.median(resample1)-np.median(resample2)
    return x
```

## Bootstrapping algorithm for the difference between proportions of people who make less than 15000 compared to those who make over 100000.
### Samples the entire cleaned data set twice and takes lower than 15000 from the first set and above 100000 from the second dataset, then takes the difference 

```python
def  prop_diff_bootstrap(sample_data, number_of_resamples):
    x = [0]* number_of_resamples 
    for i in  range(number_of_resamples):
        resample1 = [
            sample_data[j]
            for j
            in np.random.randint(len(sample_data), size=len(sample_data))
        ]
        resample2 = [
            sample_data[j]
            for j
            in np.random.randint(len(sample_data), size=len(sample_data))
        ]
        under15 = 0;
        over100 = 0;
        for j in range(len(sample_data)):
            if(resample1[j]<15000):
                under15 += 1
            if(resample2[j]>100000):
                over100 += 1
        p15 = under15/485.0   # 485.0 is needed to make the variable a decimal
        p100 = over100/485.0  # 485.0 is the number of rows in the cleaned dataframe
        x[i] = p15-p100
    return x
```

## The bootstrapping algorithm for difference of averages of datasets
### Will take Florida ages and compare them to NY ages

```python
def avg_diff_bootstrap(sample_data1, sample_data2, number_of_resamples):
    x = [0]*number_of_resamples
    for i in range(number_of_resamples):
        resample1 = [
            sample_data1[j]
            for j 
            in np.random.randint(len(sample_data1), size=len(sample_data1))
        ]
        resample2 = [
            sample_data2[j]
            for j 
            in np.random.randint(len(sample_data2), size=len(sample_data2))
        ]
        x[i] = np.mean(resample1)-np.mean(resample2)
    return x
```

## Bootstrapping algorithm for finding the difference of standard deviation between two sets
### Will be used to sample Florida and NY ages and compare the standard deviation

```python
def std_diff_bootstrap(sample_data1, sample_data2, number_of_resamples):
    x = [0]*number_of_resamples
    for i in range(number_of_resamples):
        resample1 = [
            sample_data1[j]
            for j 
            in np.random.randint(len(sample_data1), size=len(sample_data1))
        ]
        resample2 = [
            sample_data2[j]
            for j 
            in np.random.randint(len(sample_data2), size=len(sample_data2))
        ]
        x[i] = np.std(resample1)-np.std(resample2)
    return x
```

# Resets all of the indices of the dataframes
## This will allow for the bootstrapping algorithm to more easily grab any elements in each of the dataframes

```python
cali_tfi.reset_index(inplace=True)
tex_tfi.reset_index(inplace=True)
range_tfi.reset_index(inplace=True)
florida_age_avg.reset_index(inplace=True)
ny_age_avg.reset_index(inplace=True)
florida_age_std.reset_index(inplace=True)
ny_age_std.reset_index(inplace=True)
```

# The data from the empirical distribution shows that California definitely has a higher median total family income
## It is interesting to note that there was a sampling which had Texas as the higher median. This was unexpected, yet did not change the overall outcome

```python
median_boot = median_diff_bootstrap(cali_tfi['total_family_income'],tex_tfi['total_family_income'],100)
plt.hist(median_boot)
plt.show()
```

## The confidence interval of this empirical distribution shows a significant difference between the median family incomes of California and Texas, with California having a higher median

```python
ci_median = [np.percentile(median_boot,2.5), np.percentile(median_boot,97.5)]
print "The 95% confidence interval for difference of median family income of California vs Texas is: ", ci_median
```

# This empirical distribution is very normally distributed and shows a higher proportion of people are making less than 15000 in total family income

```python
prop_boot = prop_diff_bootstrap(range_tfi['total_family_income'], 100)
plt.hist(prop_boot)
plt.show()
```

## The confidence interval of 95% is close to zero, but not past zero, meaning that there is a significant enough difference between the proportion of people who make under 15000 in family income and people who make over 100000, with people who make under 15000 having a higher proportion.

```python
ci_prop = [np.percentile(prop_boot,2.5), np.percentile(prop_boot,97.5)]
print "The 95% confidence interval for difference in proportion between those who make <15000 in family income vs those who make >100000 is: ", ci_prop
```

# The bootstrap for the average age between Florida and NY shows Florida as having a generally higher average age

```python
ci_avg = [np.percentile(avg_boot,2.5), np.percentile(avg_boot,97.5)]
print "The 95% confidence interval for difference of average age of Florida vs New York is: ", ci_avg
```
