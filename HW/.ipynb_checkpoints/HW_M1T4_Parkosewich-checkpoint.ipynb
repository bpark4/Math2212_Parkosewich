# Problem 2

You can handle ties in rank() from pandas by setting 'method' to:

average: takes average rank of tied rows and applies it to all

min: takes smallest rank of tied rows and applies it to all 

max: takes largest rank of tied rows and applies to all

first: ranks based on order of appearence

dense: takes smalles rank of tied rows and applies to all, increments by 1 for each grouping

# Problem 3

```python
#This code shows a few warnings, but still completes the dataframe if you scroll past them

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

cen = pd.read_csv('census.csv')
cen_nnull = cen[ (cen['census_year'].notnull()) & (cen['state_fips_code'].notnull()) & (cen['total_family_income'].notnull()) & (cen['age'].notnull()) & (cen['sex'].notnull()) & (cen['race_general'].notnull()) & (cen['marital_status'].notnull()) & (cen['total_personal_income'].notnull())]

cen_nnull['Sqrt_TFI'] = cen_nnull['total_family_income'].apply(np.sqrt)
cen_nnull['2020_TFI'] = cen_nnull['total_family_income'].apply(lambda x: x*(.988**20))
cen_nnull['Birth_year'] = cen_nnull['age'].apply(lambda x: 2000-x)
cen_nnull
```
