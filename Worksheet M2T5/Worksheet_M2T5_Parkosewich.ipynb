```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

file = open('./adult.data', 'r')
def chr_int(a):
    if a.isdigit(): return int(a)
    else: return 0

data = []
for line in file:
    data1 = line.split(', ')
    if len(data1) == 15:
        data.append([
            chr_int(data1[0]), data1[1],
            chr_int(data1[2]), data1[3],
            chr_int(data1[4]), data1[5],
            data1[6], data1[7], data1[8],
            data1[9], chr_int(data1[10]),
            chr_int(data1[11]),
            chr_int(data1[12]),
            data1[13], data1[14].strip()
        ])
adult = pd.DataFrame(
    data,
    columns = [
        'age', 'type_employer', 'fnlwgt',
        'education', 'education_num', 'marital',
        'occupation', 'relationship','race',
        'sex', 'capital_gain', 'capital_loss',
        'hr_per_week', 'country', 'income'
    ]

)

men = adult[adult['sex'] == "Male"]
women = adult[adult['sex'] == "Female"]
men_high_income = adult[(adult['sex']=="Male") &(adult['income'] == ">50K")]
women_high_income = adult[(adult['sex']=="Female") &(adult['income'] == ">50K")]
men_low_income = adult[(adult['sex']=="Male") &(adult['income'] == "<=50K")]
women_low_income = adult[(adult['sex']=="Female") &(adult['income'] == "<=50K")]

def IQR(x):
    return x.quantile(.75) - x.quantile(.25)
```

# Question 4

The first part of the code, count_emp, is the variable used to store the catagorical variable count. adult is the dataframe with which to get the catagorical date. groupby() takes a column name, in this case type_employer, to groupby all possible options. size() gets the number of elemenents in each subsection of the column type_employer. Then the command count_emp prints the data in the variable count_emp, which is a list of all of the possible options within the column type_employer and their frequency within the dataframe. The last command builds and prints a bar graph of all of the possible options and their counts.

```python
count_emp = adult.groupby(['type_employer']).size()
count_emp
```

```python
count_emp.plot(kind="barh")
```
